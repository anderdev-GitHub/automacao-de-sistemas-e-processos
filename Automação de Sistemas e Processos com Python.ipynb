# Automação de Sistemas e Processos com Python

### Desafio:

Para controle de custos, todos os dias, seu chefe pede um relatório com todas as compras de mercadorias da empresa.
O seu trabalho, como analista, é enviar um e-mail para ele, assim que começar a trabalhar, com o total gasto, a quantidade de produtos compradas e o preço médio dos produtos.

E-mail do seu chefe: para o nosso exercício, coloque um e-mail seu como sendo o e-mail do seu chefe<br>
Link de acesso ao sistema da empresa: https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
import pyautogui
import time

# pyautogui.click -> clicar com o mouse
# pyautogui.write -> escrever um texto
# pyautogui.press -> apertar uma tecla
# pyautogui.hotkey -> apertar uma combinação de teclas

pyautogui.PAUSE = 1

# Passo 1: Acessar o sistema da empresa
pyautogui.hotkey ("ctrl", "t")
pyautogui.write ("https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema")
pyautogui.press ("enter")

time.sleep(5)

# Passo 2: Fazer o login no sistema

pyautogui.click(x=679, y=375)
pyautogui.write("anderdev")

pyautogui.click(x=633, y=459)
pyautogui.write("123")

pyautogui.click(x=692, y=532)

time.sleep(5)
# Passo 3: Baixar a base de dados

pyautogui.click(x=1264, y=360)
pyautogui.click(x=1033, y=589)

time.sleep(3)

pyautogui.click(x=1015, y=154)

time.sleep(3)

pyautogui.click(x=66, y=198)

time.sleep(3)

pyautogui.click(x=66, y=198)
pyautogui.click(x=503, y=453)
pyautogui.click(x=1156, y=50)

time.sleep(5)
```

```python
# Passo 4: Calcular os indicadores -> importando com a biblioteca pandas -> apelidar com comando "as" a biblioteca 
import pandas as pd

# importar a base de dados -> adicionar letra "r" antes do camninho para que o camninho seja comprendido corretamente pelo python
tabela = pd.read_csv(r"C:\Users\ander\Downloads\Compras.csv", sep=";") # sep="" -> separador de colunas da tabela
display(tabela)

# calculo dos indicadores
# total gasto -> somar a coluna valor final
total_gasto = tabela["ValorFinal"].sum()

# quantidade -> somar a coluna quantidade
quantidade = tabela["Quantidade"].sum()

# preço médio -> total gasto / quantidade
preco_medio = total_gasto / quantidade

print(total_gasto)
print(quantidade)
print(preco_medio)
```

```python
import pyperclip
# Passo 5: Enviar o email para a diretoria/para o chefe

# entrar no email
pyautogui.hotkey ("ctrl", "t")
pyautogui.write ("https://mail.google.com/mail/u/0/#inbox")
pyautogui.press ("enter")

time.sleep(5)

# clicar em escrever

pyautogui.click(x=83, y=198)

# preencher o email

pyautogui.write ("andersondiasleite@icloud.com")
pyautogui.press ("tab") # seleciona o email

pyautogui.press ("tab") # passar pro campo do assunto
pyperclip.copy ("Relatório de Compras")
pyautogui.hotkey("ctrl", "v")

pyautogui.press ("tab") # passar para o campo do corpo do email

# a letra "f" antes do texto formata o texto para que receba o valor de variáves dentro do mesmo -> nota: para que reconheça o valor da varável é necessario incluir detro de chaves.
# "":,.2f" -> codigo de formartação de separador de milhar a logica: ":" -> dizer que voce quer formatar "," -> separador de milhar ".2f" seria o numero de quantidades de casas decimais e o "2" é a quantidade.

texto = f"""
Prezados,

Segue relatório de compras

Total gasto: R${total_gasto:,.2f}
Quantidade de produtos: {quantidade:,}
Preço médio: R${preco_medio:,.2f}

Quaisquer dúvidas ou maiores esclarecimentos, retornar a esse email.

Favor acusar recebimento,

Att: Anderson Leite.

"""
pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")

# enviar
pyautogui.hotkey("ctrl", "enter")
```
